# Plotting and analyzing a model of a central heating system

(``Annex60.Experimental.Benchmarks.CentralHeatingSystem.System1``)

## Setup

First, import the Python modules and establish some settings for this IPython notebook:

```python
import matplotlib.pyplot as plt
import pandas
from modelicares import SimRes
%matplotlib inline
%precision 4
pandas.set_option('display.max_rows', 5)
pandas.set_option('display.max_columns', 8)
pandas.set_option('display.float_format', lambda x: '%.2f' % x)
```

Then load the simulation result:

```python
sim = SimRes('System1.mat')
```

## Plots

We'll plot some temperatures:

```python
sim.plot(['climate.T', 'room.heaCap.T', 'room.port_sou.T'], 
         legends1=["Exterior ambient", "Interior ambient", "Radiator"],
         ylabel1="Temperature", 
         title="Temperatures in " + sim.fbase + " over a year");
```

Let's look closer at the interior ambient temperature:

```python
ax, __ = sim.plot('room.heaCap.T', color=['g'], dashes=[(None, None)],
                  legends1=None,
                  ylabel1="Temperature", 
                  title="Interior ambient temperature in " + sim.fbase + " over a year");

# Show the min, max, and mean.
from modelicares import add_hlines
T_int = sim['room.heaCap.T']
add_hlines(ax, [T_int.min(), T_int.max(), T_int.mean()], 
               ["min", "max", "mean"], color='k', linestyle=":")
```

The maximum seems strange! When is the temperature greater than 296.5 K?

```python
T_int.times()[T_int.values() > 296.5]
```

The maximum is at time zero, so it is hidden by the axis. The initial value is 296.65 K:

```python
T_int.IV()
```

Now let's create a scatterplot that relates the radiator power to the exterior temperature:

```python
sim.plot(xname='climate.T',ynames1='room.port_sou.Q_flow', 
         dashes=[(1e-300, 1)],
         xlabel="Exterior temperature", marker=['.'],
         ylabel1="Radiator power", legends1=None,
         title="Radiator power vs exterior temperature");
```

## Analysis

Display some statistics:

```python
T_ext = sim['climate.T']
P_rad = sim['room.port_sou.Q_flow']
print("Minimum room temperature: %.1f %s" % (T_int.min(), T_int.unit))
print("Maximum room temperature: %.1f %s" % (T_int.max(), T_int.unit))
print("Average exterior temperature: %.1f %s" % (T_ext.mean(), T_ext.unit))
print("Average power from radiator: %.1f %s" % (P_rad.mean(), P_rad.unit))
```

Get the exterior temperature at the beginning of the 30th day:

```python
T_ext.values(t=30*86400)
```
